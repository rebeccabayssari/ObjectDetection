```python
import ultralytics.engine.exporter
from pathlib import Path

# Patch the _add_tflite_metadata function
def patched_add_tflite_metadata(self, file):
    tmp_file = Path(file).parent / "temp_meta.txt"
    with open(tmp_file, "w", encoding="utf-8") as f:  # Force UTF-8 encoding
        f.write(str(self.metadata))

ultralytics.engine.exporter.Exporter._add_tflite_metadata = patched_add_tflite_metadata
```
